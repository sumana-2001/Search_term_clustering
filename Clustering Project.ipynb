# KMEANS on direct data

```python
df_original = pd.read_excel('data v2.xlsx')
```

```python
df = df_original.iloc[:, 1:]

#replacing nulls with mean along a row instead of a column
def replace_null_with_mean(row):
    mean = row.mean(skipna=True)
    return row.fillna(mean)

df = df.apply(replace_null_with_mean, axis=1)
```

```python
df.isnull().values.any()
```

```python
cols = df.columns
from sklearn.preprocessing import MinMaxScaler
ms = MinMaxScaler()
X = ms.fit_transform(df)
X = pd.DataFrame(X, columns=[cols])
X.head()
```

```python
df.head()
```

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from kneed import KneeLocator

# WCSS = within-cluster sum of squares 
#The WCSS measures the sum of squared distances between data points and their respective cluster centers.

wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# Use the KneeLocator to find the optimal number of clusters
knee = KneeLocator(range(1, 11), wcss, curve='convex', direction='decreasing')

# Plot the elbow curve with the knee point
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.title('Elbow Method for Optimal k')
plt.vlines(knee.elbow, plt.ylim()[0], plt.ylim()[1], linestyles='dashed', colors='red')
plt.annotate(f'Optimal k = {knee.elbow}', xy=(knee.elbow, plt.ylim()[1]), xytext=(knee.elbow + 1, plt.ylim()[1] * 0.9),
             arrowprops=dict(arrowstyle='->'))
plt.show()

# The optimal number of clusters
optimal_k = knee.elbow
print("Optimal number of clusters:", optimal_k)
```
